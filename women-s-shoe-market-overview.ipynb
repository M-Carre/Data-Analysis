# Cleaning plan

Here's a cleaning plan for the dataset:

**Remove Entirely Empty Columns:**
* 'prices.condition', 'prices.dateAdded', 'prices.merchant', and 'prices.returnPolicy' are 100% empty. These columns should be removed as they provide no information.

**Handle Columns with High Percentage of Missing Values:**

**Data Type Conversions:**

**Duplicate Handling:**
* Check for and remove any duplicate entries based on all columns or a subset of key columns.

**Outlier Detection and Handling:**

**Consistency Checks:**

**Create Derived Features:**

**Handle Multi-value Fields:**
* For columns like 'colors' that may contain multiple values, consider splitting them into separate columns or creating a standardized list format.



```python
import pandas as pd
import numpy as np


# Remove entirely empty columns
df = df.drop(['prices.condition', 'prices.dateAdded', 'prices.merchant', 'prices.returnPolicy', 'asins'], axis=1)

# Handle columns with high percentage of missing values
df = df.drop('prices.shipping', axis=1)
df['dimension'] = df['dimension'].fillna('Unknown')

# Address columns with moderate missing values
df['manufacturer'] = df['manufacturer'].fillna('Unknown')
df['manufacturerNumber'] = df['manufacturerNumber'].fillna('Unknown')
df['prices.isSale'] = df['prices.isSale'].fillna(False)

# Handle Image URLs
df['imageURLs'] = df['imageURLs'].fillna('No image available')

# Convert date columns to datetime
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])

# Ensure price columns are float
df['prices.amountMax'] = pd.to_numeric(df['prices.amountMax'], errors='coerce')
df['prices.amountMin'] = pd.to_numeric(df['prices.amountMin'], errors='coerce')

# Create derived features
df['price_range'] = df['prices.amountMax'] - df['prices.amountMin']
df['year_added'] = df['dateAdded'].dt.year
df['month_added'] = df['dateAdded'].dt.month

# Display the first few rows and info of the cleaned dataset
print(df.head())
print(df.info())

# Check for any remaining missing values
print(df.isnull().sum())

print("Cleaning process completed.")
```

## Are there any significant price differences between brands?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Get the top 10 brands by frequency
top_brands = df['brand'].value_counts().nlargest(10).index

# Filter the dataframe to include only the top 10 brands
df_top_brands = df[df['brand'].isin(top_brands)]

# Calculate average price for each top brand
brand_avg_prices = df_top_brands.groupby('brand')['prices.amountMax'].mean().sort_values(ascending=False)

# Set up the plot style
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("viridis")

# Create the bar plot
ax = sns.barplot(x=brand_avg_prices.index, y=brand_avg_prices.values)

# Customize the plot
plt.title("Average Prices of Top 10 Women's Shoe Brands", fontsize=16)
plt.xlabel("Brand", fontsize=12)
plt.ylabel("Average Price (USD)", fontsize=12)
plt.xticks(rotation=45, ha='right')

# Add value labels on top of each bar
for i, v in enumerate(brand_avg_prices.values):
    ax.text(i, v + 1, f'${v:.2f}', ha='center', va='bottom')

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('top_brands_average_prices.png', dpi=300, bbox_inches='tight')

print("Plot saved as 'top_brands_average_prices.png'")

# Display some statistics
print("\
Average Prices for Top 10 Brands:")
print(brand_avg_prices)

# Calculate overall average price
overall_avg_price = df['prices.amountMax'].mean()
print(f"\
Overall Average Price: ${overall_avg_price:.2f}")

# Perform one-way ANOVA to test for significant differences
from scipy import stats

brand_groups = [group['prices.amountMax'].dropna() for name, group in df_top_brands.groupby('brand')]
f_statistic, p_value = stats.f_oneway(*brand_groups)

print(f"\
One-way ANOVA results:")
print(f"F-statistic: {f_statistic}")
print(f"p-value: {p_value}")
```

## Are there any brands that dominate certain categories?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Function to get top brands for a specific category
def get_top_brands_for_category(category, n=5):
    category_df = df[df['categories'].str.contains(category, case=False, na=False)]
    return category_df['brand'].value_counts().head(n)

# List of interesting categories to analyze
interesting_categories = [
    'Athletic', 'Sandals', 'Boots', 'Flats', 'Heels', 'Slippers',
    'Loafers', 'Clogs', 'Mules', 'Occupational & Safety'
]

# Analyze top brands for each category
results = {}
for category in interesting_categories:
    top_brands = get_top_brands_for_category(category)
    if not top_brands.empty:
        results[category] = top_brands

# Print results
for category, brands in results.items():
    print(f"\nTop 5 brands in {category} category:")
    print(brands)
    print()

# Visualize the results for a few selected categories
selected_categories = ['Athletic', 'Sandals', 'Boots', 'Heels']

fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle('Top Brands in Selected Categories', fontsize=20)

for i, category in enumerate(selected_categories):
    ax = axes[i // 2, i % 2]
    brands = results[category]
    sns.barplot(x=brands.index, y=brands.values, ax=ax)
    ax.set_title(f'Top 5 Brands in {category} Category')
    ax.set_xlabel('Brand')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('top_brands_categories.png')
print("Plot saved as 'top_brands_categories.png'")

# Calculate and print the percentage market share for the top brand in each category
print("\nMarket share of top brand in each category:")
for category, brands in results.items():
    total_shoes = df[df['categories'].str.contains(category, case=False, na=False)].shape[0]
    top_brand = brands.index[0]
    top_brand_count = brands.iloc[0]
    market_share = (top_brand_count / total_shoes) * 100
    print(f"{category}: {top_brand} ({market_share:.2f}%)")

# Display the first few rows of the dataframe
print("\nFirst few rows of the dataframe:")
print(df.head())

# Display information about the dataframe
print("\nDataframe info:")
print(df.info())
```

**Key observations:**

* Strong category dominance: In many categories, a single brand holds a significant market share. For example, Propet dominates the Athletic and Sandals categories, while Brinley Co. leads in Boots and Flats.

* Niche specialization: Some brands appear to specialize in specific categories. For instance, minnetonka is prominent in the Slippers category but doesn't appear in the top 5 for other categories.

* Limited data in some categories: For categories like Heels, Loafers, Clogs, Mules, and Occupational & Safety, there seems to be limited data, resulting in 100% market share for a single brand. This suggests that the dataset might not fully represent these categories.

* Versatile brands: Some brands, like Propet and Brinley Co., appear in multiple categories, indicating a diverse product range.

* Unexpected findings: It's surprising to see Fila dominating categories like Clogs, Mules, and Occupational & Safety shoes, which are not typically associated with the brand. This could be due to data limitations or miscategorization.

## Are there any price differences based on color?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


# Function to extract colors from the 'colors' column
def extract_colors(color_string):
    if pd.isna(color_string):
        return []
    return [color.strip().lower() for color in color_string.split(',')]

# Extract all colors and create a new dataframe with color and price
color_price_data = []
for _, row in df.iterrows():
    colors = extract_colors(row['colors'])
    for color in colors:
        color_price_data.append({'color': color, 'price': row['prices.amountMax']})

color_price_df = pd.DataFrame(color_price_data)

# Calculate average price for each color
avg_price_by_color = color_price_df.groupby('color')['price'].mean().sort_values(ascending=False)

# Get top 15 colors by frequency
top_15_colors = color_price_df['color'].value_counts().head(15).index

# Filter average prices for top 15 colors
top_15_avg_prices = avg_price_by_color[top_15_colors].sort_values(ascending=False)

# Create a bar plot for top 15 colors by average price
plt.figure(figsize=(12, 6))
sns.barplot(x=top_15_avg_prices.index, y=top_15_avg_prices.values)
plt.title("Average Price of Women's Shoes by Color (Top 15 Colors)")
plt.xlabel('Color')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('avg_price_by_color.png')

print("Average price of women's shoes by color (Top 15 colors):")
print(top_15_avg_prices)

# Calculate overall average price
overall_avg_price = color_price_df['price'].mean()
print(f"\
Overall average price: ${overall_avg_price:.2f}")

# Calculate price difference from overall average
price_diff = top_15_avg_prices - overall_avg_price
print("\
Price difference from overall average:")
print(price_diff)

# Calculate percentage difference from overall average
percentage_diff = (price_diff / overall_avg_price) * 100
print("\
Percentage difference from overall average:")
print(percentage_diff)

print("\
Plot saved as 'avg_price_by_color.png'")

# Perform statistical test (ANOVA) to check if the differences are significant
# Prepare data for ANOVA
color_groups = [group for name, group in color_price_df[color_price_df['color'].isin(top_15_colors)].groupby('color')['price']]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(*color_groups)

print(f"\
One-way ANOVA results:")
print(f"F-statistic: {f_statistic}")
print(f"p-value: {p_value}")

if p_value < 0.05:
    print("The differences in prices between colors are statistically significant.")
else:
    print("The differences in prices between colors are not statistically significant.")
```

**Key Observations:**

* Price Range: There's a substantial price range among the top 15 colors, with white shoes being the most expensive (83.81 USD on average) and grey shoes being the least expensive (29.67 USD on average).

* Premium Colors: White, bronze, bone, and royal blue seem to command premium prices, all being more than 10% above the overall average price.

* Budget-Friendly Colors: Grey, taupe, and brown are notably less expensive than the average, with prices 23% to 53% below the overall average.

* Neutral Colors: Interestingly, neutral colors span both ends of the price spectrum. White and off-white are among the most expensive, while grey and taupe are among the least expensive.

* Popular vs. Expensive: Some of the most popular colors (like black and blue) are actually priced below the average, suggesting that popularity doesn't necessarily correlate with higher prices.

* Statistical Significance: The ANOVA test confirms that these price differences are not due to random chance, but represent real pricing trends in the market.

**Possible Explanations:**

* Material Quality: Certain colors might be associated with higher-quality materials or more complex manufacturing processes, leading to higher prices.

* Seasonality: Some colors might be more in-demand during certain seasons, potentially affecting their price.

* Brand Association: Particular colors might be more commonly used by luxury or high-end brands, driving up their average price.
 
* Versatility: Colors perceived as more versatile (like white or black) might be priced differently based on their perceived value to consumers.
 
* Production Costs: Some colors might be more expensive to produce consistently or might require more expensive dyes or treatments.
